```python
def do_some_prep(df):
    cols = df.columns
    df.rename(columns = {cols[2]: 'Num'}, inplace = True)
    n = {'Five': 5, 'Four': 4, 'Three': 3}
    df['Num'].replace(n, inplace = True)
    
    '''
    ## these a, b, c are used in the checks but are extra here
    '''
    a = df.Name=='Name'
    # df.loc[a]
    b = df.Name.isnull()
    # df.loc[b]
    c = df[~(a | b)]
    # d = c['Number\nof\nMembers'].replace(n)
    c.Num.sum()
    
    return df

df = do_some_prep(df)
```

```python
def sched_squeeze(df):
    d = pd.DataFrame(columns=df.columns)
    
    indx = 0
    r = df.iloc[indx]
    name = r.Name.replace('\n', '')
    area = r.Area
    num = r.Num
    for i in range(1, len(df)):
        r = df.iloc[i]
        n = r.Name
        if type(n) is float:
            area = f'{area}\n{r.Area}'
        elif r.Name == 'Name':
#             df.drop()
            continue
        else:
#             df.iloc[indx].Area = area
            d.loc[indx] = [name, area, num]
#             dic = {'Name': name, 'Area': area, 'Num': num}
#             a = pd.DataFrame(dic)
#             print(a)
#             d.append(a, ignore_index=True)
#             print(dic)
#             d.append({'Name': name, 'Area': area, 'Num': num}, ignore_index=True)
            indx = indx +1
            name = n.replace('\n', '')
            area = r.Area
            num = r.Num
    d.loc[indx] = [name, area, num]
    return d
```
